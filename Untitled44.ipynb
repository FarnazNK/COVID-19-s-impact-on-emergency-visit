```python
# Lists to store dataframes
edv_list = [ED_Volumes_2019, ED_Volumes_2020]
year_list = [2019, 2020]


# Selecting columns, changing column names, and removing year from Date_of_visit column
for i, edv in enumerate(edv_list):
    edv_sub = edv[['Date_of_visit','P.E.I.','N.S.','Que.','Ont.','Man.','Sask.','Alta.','B.C.']]
    edv_sub = edv_sub.rename(columns = {'Ont.':f'Ont.{year_list[i]}', 'Que.':f'Que.{year_list[i]}',
                                        'Alta.':f'Alta.{year_list[i]}','B.C.':f'B.C.{year_list[i]}',
                                        'Man.':f'Man.{year_list[i]}', 'Sask.':f'Sask.{year_list[i]}',
                                        'N.S.':f'N.S.{year_list[i]}','P.E.I.':f'P.E.I.{year_list[i]}',}, 
                              inplace = False)
    edv_sub['Date_of_visit'] = edv_sub['Date_of_visit'].dt.strftime('%B-%d')
    if i == 0:
        edv_merged = edv_sub
    else:
        edv_merged = pd.merge(edv_merged, edv_sub, on='Date_of_visit', how='outer')

# Changing Date_of_visit format
edv_merged
```

**A line graph comparing the daily moving average of emergency department visits from 2019 to 2020**



```python
# List of provinces
provinces = ['P.E.I.', 'N.S.', 'Que.', 'Ont.', 'Man.', 'Sask.', 'Alta.', 'B.C.']

# Loop through provinces
for province in provinces:
    # Extracting data for each province
    df = edv_merged[['Date_of_visit', '{}2019'.format(province), '{}2020'.format(province)]]
    
    # Plotting the data for each province
    df.plot(x='Date_of_visit', y=['{}2019'.format(province), '{}2020'.format(province)], figsize=(18,10), fontsize=17)
    plt.title('ED Visits Comparison ({} - 2019 vs 2020)'.format(province), fontsize=20)
    plt.show()
```

As the graphs shows, at all provinence, the number of visits were consistent in 2019. In 2020, there is a sharp decrease in all provinence, but

```python
# List of years
years = [2019, 2020]

# Initialize empty list to store intermediate dataframes
df_list = []

# Loop through years
for year in years:
    # Selecting columns
    ED_Volumes = eval("ED_Volumes_{}".format(year))[['Date_of_visit', 'Total']]
    ED_Volumes = ED_Volumes.rename(columns = {'Total':'Total.{}'.format(year)}, inplace = False)

    # Extracting month and day from timestamp and concatenating them together
    ED_Volumes['Date_of_visit'] = ED_Volumes['Date_of_visit'].dt.strftime("%B-%d")
    
    # Adding moving average to the dataset
    ED_Volumes['MAVG_{}'.format(year)] = ED_Volumes.expanding(min_periods=7).mean()
    
    # Append intermediate dataframe to the list
    df_list.append(ED_Volumes)

# Merging datasets                  
ED_Volumes_19t20_t = df_list[0].merge(df_list[1], on='Date_of_visit', how='outer')

# Line plot of moving average of ED visits by day, comparing 2019 vs 2020
ED_Volumes_19t20_t.plot(x='Date_of_visit', y=['MAVG_2019','MAVG_2020'], figsize=(18,10), fontsize=17)
plt.title('Seven-day moving average ED visits comparison (2019 vs 2020)', fontsize=20)
```

As the data shows, the number of visits in 2019 was more consistent compared to 2020. After March 21st, there was a sharp decrease in the number of visits per week.

```python
#Creates two dataframes ED_IA_2019 and ED_IA_2020, adding a "Year" column to each with the respective year.
ED_IA_2019['Year'] = '2019'
ED_IA_2020 ['Year'] = '2020'
df1 = [ED_IA_2019, ED_IA_2020]

#Combine these two dataframes into a new one edia_1920 using the pd.concat function, and resets the index
ED_IA_1920 = pd.concat(df1)
ED_IA_1920 = ED_IA_1920.reset_index()


#Change the data type of the "Year" columns in both ED_IA_2019 and ED_IA_2020 to integer using the astype method
ED_IA_2019['Year'] = ED_IA_2019['Year'].astype(int)
ED_IA_2020['Year'] = ED_IA_2020['Year'].astype(int)

#Create a bar plot, with each bar representing the median time to physician initial assessment for a given province/territory, 
#broken down by year, for the months of March, April, May, and June.
months = ['March', 'April', 'May', 'June']

fig, axs = plt.subplots(2, 2, figsize=(22,15))
fig.suptitle('Median time to physician initial assessment', fontsize=20)

for i, month in enumerate(months):
    sns.barplot(x="Province/territory", y=month, hue="Year", data=ED_IA_1920, ax=axs[i//2, i%2])

plt.show()
```
